## Item-Item CF


```python
# data cleaning
cf_data = data[['city', 'budget_level', 'culture', 'adventure', 'nature', 'beaches', 'nightlife', 'cuisine', 'wellness', 'urban', 'seclusion']]

cf_data.set_index('city', inplace=True)

cf_data.head()
```

```python
onehot_X = pd.get_dummies(cf_data, drop_first=True, dtype = int)
onehot_X.head()
```

```python
city_df = onehot_X.T
city_df
```
